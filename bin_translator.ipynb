```python
import numpy as np
from bitstring import ConstBitStream
import matplotlib.pyplot as plt
from numpy.fft import fft, ifft
from bitstring import Bits

# test_fn = "./control/DA1_20.BIN"
list_fn = "./down/fnlist.txt"
```

```python
fn_list_arr = np.loadtxt(list_fn, dtype=str)
chunks = len(fn_list_arr)

samples = int(524288/16)

int_arr = np.zeros([chunks, samples])

for fi, fn in enumerate(fn_list_arr):
    b = ConstBitStream(filename="./down/"+fn)
    for i in range(samples):
        b1 = b.read(4).bin
        b.read(4)
        b2 = b.read(8).bin
        int_arr[fi, i] =  Bits(bin=b2+b1).int
```

```python
np.save("down_ts.npy", int_arr)
```
